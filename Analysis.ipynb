```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
```

```python
df['Total_Power_consumed']=df['Power Galley 2 (MW)']+ df['Power Service (MW)']+df['HVAC Chiller 1 Power (MW)']+df ['HVAC Chiller 2 Power (MW)']+df['HVAC Chiller 3 Power (MW)']+ df['Scrubber Power (MW)']+df['Port Side Propulsion Power (MW)']+df['Starboard Side Propulsion Power (MW)']+ df['Bow Thruster 1 Power (MW)']+df['Bow Thruster 2 Power (MW)']+ df['Bow Thruster 3 Power (MW)']+df['Stern Thruster 1 Power (MW)']+df[ 'Stern Thruster 2 Power (MW)']
```

```python
df['Power_Produced']=df['Diesel Generator 1 Power (MW)']+df ['Diesel Generator 2 Power (MW)']+df['Diesel Generator 3 Power (MW)']+ df['Diesel Generator 4 Power (MW)']
```

```python
df_vessel1 = df[df['Vessel Name'] == 'Vessel 1']
df_vessel2 = df[df['Vessel Name'] == 'Vessel 2']
```

#### I have calculated the total power consumed and total power produced in each vessel. Then I have done an analysis of how much power each vessel consumes and generates. The time fields have been replaced with a linear list.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=[i for i in range(len(df_vessel1))],y=df_vessel1['Power_Produced'])

plt.title('Total Power Produced in vessel 1')
plt.legend(title='Vessel 1')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=[i for i in range(len(df_vessel2))],y=df_vessel2['Power_Produced'])

plt.title('Total Power Produced in vessel 2')
plt.legend(title='Vessel 2')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=[i for i in range(len(df_vessel2))],y=df_vessel2['Total_Power_consumed'])

plt.title('Total Power consumed in vessel 2')
plt.legend(title='Vessel 2')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=[i for i in range(len(df_vessel2))],y=df_vessel2['Total_Power_consumed'])

plt.title('Total Power consumed in vessel 2')
plt.legend(title='Vessel 2')
plt.grid(True)
plt.show()
```

#### An analysis of how much power each HVAC Chiller is consuming in each vessel.

```python
hvac_scrubber_data = df.groupby("Vessel Name")[["HVAC Chiller 1 Power (MW)", "HVAC Chiller 2 Power (MW)", "HVAC Chiller 3 Power (MW)", "Scrubber Power (MW)"]].mean().reset_index()

hvac_scrubber_data.plot(kind="bar", x="Vessel Name", stacked=True, figsize=(10, 6))
plt.title("Average HVAC and Scrubber Power Consumption by Vessel")
plt.xlabel("Vessel Name")
plt.ylabel("Power Consumption (MW)")
plt.legend(title="System Components")
plt.show()
```

```python
# Sum auxiliary fuel flow for each vessel
df["Total Auxiliary Fuel Flow"] = df[["Boiler 1 Fuel Flow Rate (L/h)", "Boiler 2 Fuel Flow Rate (L/h)", "Incinerator 1 Fuel Flow Rate (L/h)"]].sum(axis=1)
aux_fuel_data = df.groupby("Vessel Name")["Total Auxiliary Fuel Flow"].mean().reset_index()

sns.barplot(data=aux_fuel_data, x="Vessel Name", y="Total Auxiliary Fuel Flow")
plt.title("Average Auxiliary Fuel Flow by Vessel")
plt.xlabel("Vessel Name")
plt.ylabel("Total Auxiliary Fuel Flow (L/h)")
plt.show()
```

#### A scatter plot of how the fuel consumption is affected by speed with which the vessel propels through water.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Speed Through Water (knots)", y="Propulsion Power (MW)", hue="Vessel Name")
plt.title("Propulsion Power vs Speed Through Water")
plt.xlabel("Speed Through Water (knots)")
plt.ylabel("Propulsion Power (MW)")
plt.legend()
plt.show()
```

#### How much power is each thruster consuming by each vessel.

```python
thruster_data = df.groupby("Vessel Name")[["Bow Thruster 1 Power (MW)", "Bow Thruster 2 Power (MW)", "Bow Thruster 3 Power (MW)", 
                                           "Stern Thruster 1 Power (MW)", "Stern Thruster 2 Power (MW)"]].mean().reset_index()

thruster_data.plot(kind="bar", x="Vessel Name", stacked=True, figsize=(10, 6))
plt.title("Average Thruster Power Consumption by Vessel")
plt.xlabel("Vessel Name")
plt.ylabel("Power Consumption (MW)")
plt.legend(title="Thrusters")
plt.show()
```
